```python
import pandas as pd
import colorsys
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
np.set_printoptions(edgeitems=10)
```

```python
from src import load_data
```

```python
df = load_data.load_data('data/ratings')
df.head(10)
```

```python
df2 = load_data.symmetrize(df)
df
```

```python
for hue in np.linspace(0, 1, 10):
    rgb = colorsys.hsv_to_rgb(hue, 1, 1)
    print('{:.3f} {:.3f} {:.3f}'.format(*rgb))
```

```python
model = load_data.make_model(df2)
```

```python
mesh_size = 50
mesh = np.meshgrid(np.linspace(0, 1, mesh_size),
                   np.linspace(0, 1, mesh_size))
color1 = np.stack([mesh[0].flatten(),
                   np.ones(mesh_size**2),
                   np.ones(mesh_size**2)]).T
color2 = np.stack([mesh[1].flatten(),
                   np.ones(mesh_size**2),
                   np.ones(mesh_size**2)]).T
```

```python
color1.shape
```

```python
v_hsv_to_rgb = np.vectorize(lambda x: np.array(colorsys.hsv_to_rgb(*x)),
                            signature='(n)->(n)')
```

```python
X = np.concatenate([v_hsv_to_rgb(color1),
                    v_hsv_to_rgb(color2)], axis=1)
```

```python
pred = model.predict(X).reshape((mesh_size, mesh_size))
```

```python
X.round(3)
```

A graph of the predicted matchings based on the hues of the two colors.

This would be more symmetric with more trees.

```python
fig, ax = plt.subplots(2, 2, figsize=(7, 6),
                       gridspec_kw={'width_ratios': (1, 5),
                                    'height_ratios': (5, 1)})
im = ax[0, 1].pcolormesh(mesh[0], mesh[1], pred, cmap='gray')
fig.colorbar(im, ax=ax)

ax[0, 0].axis('off')
ax[1, 0].axis('off')
ax[0, 1].axis('off')
ax[1, 1].axis('off')

ax[1, 1]
```
